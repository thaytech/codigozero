```python
import pandas as pd

#pd = pandas
#df = data frame
df = pd.read_csv('estoque.csv')

display(df)
```

```python
# carregar amostra de dados
print(df.head(2))

print()
# ver as colunas do data frame
#print(df.columns)

for i in df.columns:
    print(i)
```

```python
#consultar colunas

print(df['marca'])

#consultar colunas

print(df[['marca','cor','valor']])
```

```python
# agrupar os dados

cor = df.groupby('cor')['marca'].count()

print(cor)
print('\n-----------------------------------------------------\n')
# quantos carros tem de cada marca
marca = df.groupby('marca')['marca'].count()
print(marca)
```

```python
#visualizar com for

marcas = df['marca']


for carros in marcas:
    print(f'marca:{carros}')
```

```python
#criar uma coluna nova

df['faturamento'] = df['quantidade'] * df['valor']

display(df)
```

```python
# valor total da loja - somar a coluna faturamento
total = df['faturamento'].sum()

print(f'valor total da empresa: R$ {total:.2f}reais')

estoque = df['quantidade'].sum()
print(f'quantidade de carros na loja: {estoque}')

```

```python
#filtros
#carros com menos de 10 anos de uso

novos = df[df['ano'] > 2015]
display(novos)
#carros com mais de 25 anos
velhos=df[df['ano'] <= 1999]
display(velhos)
```

```python
#filtros
#ex filtrar produtos por cor

cor= input('Qual a cor desejada:')
FiltroCor= df[ df ['cor'] == cor]
print(FiltroCor)
```
